```python
def catplot(data,x,kind,hue):#kind="violin"
    sns.catplot(data=data,kind='count',x=x,hue=hue)
    plt.show()    
```

```python
from scipy import stats
def chi_squared_test_result(col):
    
    x = data.groupby(['Revenue', col])[col].count().unstack(1).fillna(0).astype('int')
    test_results = stats.chi2_contingency(x.values)
    pval = test_results[1]
    if pval<0.05:
        print(f"P-value: {pval:.4f} - \n\nIt means we can reject the null hypothesis (at 95% CI) and conclude that Revenue is dependent on {col}.")
    else:
        print(f"P-value: {pval:.4f} - \n\nFailed to reject null hypothesis. Both Revenue is independent on {col}.")
```

# Missing value treatment

```python
data.describe()
```

```python
# All the continuous_features have the outliers
features=['BounceRates','ExitRates','Administrative_Duration','ProductRelated_Duration']
```

```python
print(data.shape)
for cols in features:
    Q1 = data[cols].quantile(0.25)
    Q3 = data[cols].quantile(0.75)
    IQR = Q3 - Q1     

    df_1 = (data[cols] >= Q1 - 1.5 * IQR) & (data[cols] <= Q3 + 1.5 *IQR)
    data_outliers_removed = data.loc[df_1]
    
print(data_outliers_removed.shape)
```

```python
def dist_plot(col):
    sns.distplot(data_outliers_removed[col])
    plt.show()
```

```python
for column in features:
    dist_plot(column)
```

# Scaling the features
### y = (x â€“ mean) / standard_deviation


```python
# We have handled the outlier, so we are going with the standard scaler
```

```python
print(continuous_features)
print(discrete_features)
```

```python
from sklearn.preprocessing import StandardScaler
feature_to_scale = [feature for feature in data.columns if feature not in discrete_features]

scaler = StandardScaler()
scaler.fit(data[feature_to_scale])
scaled_data = pd.concat([data_outliers_removed[['Administrative', 'Informational', 'ProductRelated', 'SpecialDay', 'Month', 'OperatingSystems', 'Browser', 'Region', 'TrafficType', 'VisitorType', 'Weekend', 'Revenue']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(data_outliers_removed[feature_to_scale]), columns=feature_to_scale)],
                    axis=1)
scaled_data.head()
```

```python
scaled_data.shape
```

# Label Encoding Features


```python
from sklearn.preprocessing import LabelEncoder

encoded_data=scaled_data.copy()
features_to_encode = ['Month','VisitorType','Weekend','Revenue']
label_encoder = LabelEncoder()
for col in features_to_encode:
    encoded_data[col] = label_encoder.fit_transform(scaled_data[col])
encoded_data.head()
```

```python
encoded_data.shape
```

# Handling multi collinearity  

```python
plt.figure(figsize=(20,20))
sns.heatmap(encoded_data.corr(),annot = True,cmap='coolwarm')
```

```python

# Create correlation matrix
corr_matrix = encoded_data.corr().abs()

# Select upper triangle of correlation matrix(select upper elements of kth diagonal)
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.8
positive_drop = [column for column in upper.columns if any(upper[column] >= 0.8)]
negative_drop = [column for column in upper.columns if any(upper[column] <= -0.8)]
all_drop=positive_drop+negative_drop
```

```python
print(all_drop)
```

```python
non_corr_data=encoded_data.drop(all_drop, axis=1)
```

# Splitting the dependent and independent variables


```python
from sklearn.model_selection import train_test_split
X=non_corr_data.drop(['Revenue'],axis=1)
y=non_corr_data.Revenue

X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.8,random_state = 42,stratify=y)
print("Input Training:",X_train.shape)
print("Input Test:",X_test.shape)
print("Output Training:",y_train.shape)
print("Output Test:",y_test.shape)
```

# Conclusion : With the SMOTE we are getting better recall,precsion value's for train and test

### Conclusion from different ML Models - If the aim of the model is to minimize the the undetected revenue sessions (False Negatives) and at the cost of incorrectly predicting non-revenue sessions as revenue sessions, then SMOTE model should be chosen. If the aim of the company is to just focus on the recall value, then it must go ahead with a SMOTE analysis. Further, there is a high level of overfitting in the oversampling model.
